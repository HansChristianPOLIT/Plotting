## Basics of plotting

```python
# create values
x = list(range(21)) #create list from 0 to 20
y = [val **2 for val in x]

# change line style
plt.plot(x,y, color='k',marker='^', markerfacecolor='m',markeredgecolor='k', markersize=8,linewidth=4,linestyle='dotted')

# add axis labels
plt.xlabel("Time")
plt.ylabel("Sales")

# add title
plt.title("Figure A - Sales growth", fontweight='bold', fontname='Times New Roman', size=20)

# plot a vertical line at x = ""
plt.axvline(10, color='grey')

# plot a horisontal line at y = ""
plt.axhline(200,color='grey')

# plot the points
plt.show()
```

## Subplots 

```python
# create data 
X1 = list(range(0,101))
X2 = list(range(50,101))
Y1 = [10 * x for x in X1]
Y2 = [(x-50) ** 2 for x in X2]

X3 = [x*2 for x in X1]
Y3 = [10 * x for x in X3]
X4 = [x*2 for x in X2]
Y4 = [(x-50) ** 2 for x in X4]

# basic subplots can be called with a number of rows as nrows 
# also we can adjust size to make it easier to see
# it is possible to share the x-axis & y-axis with sharex=True ...
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,8), sharex=True, sharey=True)

#Sharing with the row the x and y axis
#fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,12), sharex='row', sharey='row')

#Sharing with the columns the x and y axis
#fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(12,12), sharex='col', sharey='col')

# plot by indexing axes 
# set labels & title
ax[0,0].plot(X1, Y1)
ax[0,0].set_xlabel("Series 1 X")
ax[0,0].set_ylabel("Series 1 Y")
ax[0,0].set_title("Series 1")

ax[1,0].plot(X2, Y2)
ax[1,0].set_xlabel("Series 2 X")
ax[1,0].set_ylabel("Series 2 Y")
ax[1,0].set_title("Series 2")

ax[0,1].plot(X3, Y3)
ax[0,1].set_xlabel("Series 3 X")
ax[0,1].set_ylabel("Series 3 Y")
ax[0,1].set_title("Series 3")

ax[1,1].plot(X4, Y4)
ax[1,1].set_xlabel("Series 4 X")
ax[1,1].set_ylabel("Series 4 Y")
ax[1,1].set_title("Series 4")

plt.show()
```

### Example - Plots with different sizes

```python
#An alternative that allows you to pick different sizes for your subplots is gridspec
fig = plt.figure(constrained_layout=True, figsize=(8,8))

#Add a grid spec with 3 rows and 2 columns
qs = fig.add_gridspec(3,2)

#Add a plot at row 0, and for all columns
ax1 = fig.add_subplot(qs[0,:]) 

#Add a plot in column 1, taking up spots in the second and third row
ax2 = fig.add_subplot(qs[1:3,0]) 

#Add a plot in row 3, column 2
ax3 = fig.add_subplot(qs[2,1])

# Insert plots
ax1.plot(X1, Y1)
ax1.set_xlabel("Series 1 X")
ax1.set_ylabel("Series 1 Y")
ax1.set_title("Series 1")

ax2.plot(X2, Y2)
ax2.set_xlabel("Series 2 X")
ax2.set_ylabel("Series 2 Y")
ax2.set_title("Series 2")

ax3.plot(X3, Y3)
ax3.set_xlabel("Series 3 X")
ax3.set_ylabel("Series 3 Y")
ax3.set_title("Series 3")

plt.show()
```

## 3D Line Plots

```python
# start with a basic 2-D Spiral
import math 
import numpy as np 
import matplotlib.pyplot as plt

# Math to build the spiral
phi = np.linspace(0,math.pi*10,1000)
r = phi / math.pi
X = np.cos(phi) * r
Y = np.sin(phi) *r 

# to go from 2D to 3D, we create a z-axis
Z = r ** 2

# Create the figure
fig = plt.figure(figsize=(12,8))

# then, we add a subplot with projection of â€˜3dâ€™, and this will allow us to show it
ax = fig.add_subplot(111,projection='3d') # this is just the common way to do it in newer matplotlib

# Plot the spiral
#plt.plot(X,Y,Z)
ax.scatter(X,Y,Z) # if we want to do scatter instead
plt.show()
```
